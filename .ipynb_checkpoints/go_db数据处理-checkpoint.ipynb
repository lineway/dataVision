```python
import pymongo
from string import punctuation
```

```python
client = pymongo.MongoClient('localhost', 27017)
```

```python
db = client.go_db
```

```python
collection = db.work_detail
```

```python
for i in collection.find().limit(300):
    print(i['city'])
```

```python
for i in collection.find():
    new_data = i['city'].split('/')[1]
    print(new_data)
    collection.update_one({'_id':i['_id']}, {"$set":{'city':new_data}})
```

```python
for i in collection.find():
    print i['city']
```

```python
for i in collection.find():
    workYear = i['workYear'].split('/')[0]
    collection.update({"_id": i["_id"]}, {"$set": {"workYear": workYear}})
```

```python
for i in collection.find():
    education = i["education"].split('/')[0]
    collection.update({'_id': i['_id']}, {"$set": {"education": education}})
    print education
```

```python
for i in collection.find():
    company = i['company'].strip().split()[0]
    collection.update({"_id": i['_id']}, {"$set": {'company': company}})
    print i['company']
```

```python
for i in collection.find():
    description = i['description'].strip()
    collection.update({"_id": i["_id"]}, {"$set": {"description": description}})
    print description
```
